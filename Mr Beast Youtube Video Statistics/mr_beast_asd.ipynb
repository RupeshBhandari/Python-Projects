### Changing index
Since, the id is unqiue identifier we cou;ld use it to replace the index. 

```python
df = df.set_index('id')
df.head(2)
```

### Subsetting Data

#### Subsetting column

```python
df = df[['title', 'description', 'publishTime', 'duration_seconds','viewCount', 'likeCount', 'commentCount']]
df
```

#### Subsetting row


```python
df_sub_1 = df.loc[df['viewCount'] > 50000 ]
df_sub_1
```

### Missing Data

```python
df = df.loc[~df['viewCount'].isna()]
df
```

### Casting dtypes

```python
df.dtypes
```

```python
df['viewCount'] = df['viewCount'].astype('int')
```

Lets check for the data type again

```python
df.dtypes
```

### Convert string to date and time 

```python
pd.to_datetime(df['publishTime'])
```

### Adding a new column

```python
df['like_to_view_ratio(%)'] = df['likeCount'] / df['viewCount'] * 100
df
```

### Adding a new row

```python
df_to_append = df.tail(1)
pd.concat([df, df_to_append]).tail(2)
```
